**How much loan do people get most often?**
We can find the answer using __LoanOriginalAmount__ column.

```python
# to answer the question, we use matplotlib's hist function
bin_list = np.arange(1000, df['LoanOriginalAmount'].max() + 700, 700)
plt.hist(df['LoanOriginalAmount'], bins = bin_list);
plt.xlabel('LoanOriginalAmount ($)')
plt.ylabel('Count')
```

**From the table above you can see a colum named "Investors". I interested in how much money per investor invests in general.**

```python
# to do this we need to create a new column for investment per investor
df['investment_per_investor'] = df['LoanOriginalAmount']/df['Investors']

bin_list = 10 ** np.arange(0.86, np.log10(df['investment_per_investor'].max()) + 0.1, 0.1)
x_ticks = [10, 30, 100, 300, 1000, 3000, 10000, 30000]
x_labels = ['{}'.format(val) for val in x_ticks]

plt.hist(df['investment_per_investor'], bins = bin_list);
plt.xscale('log')
plt.xticks(x_ticks, x_labels);
plt.xlabel('InvestmentPerInvestor ($)')
plt.ylabel('Count')
```

Yeah! you can see the most popular investment between 30 and 100 dollars. But there is also big investors that can invest dozens of thousand dollars.

>**The most interesting thing I liked about the dataframe, it has got a lot of columns!** And you can take any of them as center of interest! And now I have interested in CreditGrade and ProsperScore columns.

CreditGrade column has grading for loans before 2009

- AA, A, B are good grades
- C, D are moderate grades
- E, HR or NC are low grades

ProsperScore column has grading for loans after 2009

- grading includes numbers from 1 to 11, they mean low to high grades respectively

**So, let's see which grades for loans are most often!**

```python
order_credit = df['CreditGrade'].value_counts().index
base_color = sns.color_palette()[0]
sns.countplot(data = df, x = 'CreditGrade', color = base_color, order = order_credit);
```

You can see there moderate grades for loans are most often. **What about looking through the proportions of the gradedings.**

```python
# let's first create a function that plots proportion plot. And that will be cool, you see in during the next steps
def plot_proportions(data, column):
    
    # in this part function identifies some values for plot
    column_value_counts = data[column].value_counts()
    counts_order = column_value_counts.index
    base_color = sns.color_palette()[0]
    
    # in this part proportion is calculated
    num_of_elements = data[column].shape[0]
    max_value_counts = column_value_counts.max()
    max_prop = max_value_counts/num_of_elements
    
    # function identifies some values for tick labels
    tick_props = np.arange(0, max_prop, 0.02)
    tick_names = ['{:0.2f}'.format(v) for v in tick_props]
    
    # in this part our plot will be ready
    x = sns.countplot(data = data, y = column, color = base_color, order = counts_order)
    plt.xticks(tick_props * num_of_elements, tick_names)
    plt.xlabel('proportion');
    
    return x
```

```python
# now we are ready to plot
plt.figure(figsize = (12, 5))

# let's plot proportions for two gradings at once
col_list = ['CreditGrade', 'ProsperScore']
data = df

for i in range(len(col_list)):
    
    plt.subplot(1, 2, i + 1)
    plot_proportions(data, col_list[i])
```

**Yeah, It looks like getting good grades is not easy at all!** Because the bar chart suggests that there are more proportions of moderate gradings for both before and after 2009. May be getting good credit grades depends on income range?
